```python
import pandas as pd

print("Data loading started...")

# Load the Expenditure file
# We use skiprows=3 to ignore the first 3 text-heavy rows and grab the actual headers
try:
    df_exp = pd.read_excel('expenditure.xlsx', skiprows=3) 
    print("\n--- Expenditure Data Loaded Successfully ---")
    display(df_exp.head(10))
except Exception as e:
    print("Error loading Expenditure data:", e)

# Load the Receipts (Income) file
try:
    df_rec = pd.read_excel('receipts.xlsx', skiprows=3)
    print("\n--- Receipts Data Loaded Successfully ---")
    display(df_rec.head(10))
except Exception as e:
    print("Error loading Receipts data:", e)
```

```python
print("Starting Data Cleaning...")

# 1. Remove the top 4 rows (index 0, 1, 2, 3) because they are just messy headers
df_exp_clean = df_exp.iloc[4:].copy()

# 2. Select only the most important columns (Year, Revenue, Defence, Interest, Subsidies, Capital)
columns_to_keep = df_exp_clean.columns[1:7]
df_exp_clean = df_exp_clean[columns_to_keep]

# 3. Rename these columns to clean, professional names
df_exp_clean.columns = [
    'Year', 
    'Revenue_Expenditure', 
    'Defence_Exp', 
    'Interest_Payments', 
    'Subsidies', 
    'Capital_Expenditure'
]

# 4. Reset the row numbers so they start from 0 instead of 4
df_exp_clean = df_exp_clean.reset_index(drop=True)

print("\n--- Cleaned Expenditure Data ---")
display(df_exp_clean.head())
```

```python
print("Preparing Receipts (Income) Data...")

# 1. Remove the top 4 rows (index 0, 1, 2, 3) to clear the messy headers
df_rec_clean = df_rec.iloc[4:].copy()

# 2. Reset the index so it starts from 0
df_rec_clean = df_rec_clean.reset_index(drop=True)

# 3. Display the first few rows to see the column structure
print("\n--- Raw Receipts Data (After removing top garbage rows) ---")
display(df_rec_clean.head())
```

```python
print("Cleaning Receipts Data...")

# 1. Select the meaningful columns (Index 1 to 6, skipping the garbage Unnamed: 0)
columns_to_keep_rec = df_rec_clean.columns[1:7]
df_rec_clean = df_rec_clean[columns_to_keep_rec]

# 2. Rename columns to professional, easy-to-understand names
df_rec_clean.columns = [
    'Year',
    'Total_Revenue_Receipts',
    'Tax_Revenue',
    'Direct_Tax',
    'Indirect_Tax',
    'Non_Tax_Revenue'
]

print("\n--- Cleaned Receipts Data ---")
display(df_rec_clean.head())
```

```python
print("Merging Data into a Master Dataset...")

# 1. Merge both tables using the 'Year' column
# 'inner' means it will only keep the years that are present in both tables
df_master = pd.merge(df_rec_clean, df_exp_clean, on='Year', how='inner')

# 2. Convert all financial columns to strict numbers just to be 100% safe before doing math
for col in df_master.columns:
    if col != 'Year':
        df_master[col] = pd.to_numeric(df_master[col], errors='coerce')

# 3. FEATURE ENGINEERING: Calculate the "Kami" (Deficit Gap)
# Gap = Revenue Expenditure (Kharcha) - Total Revenue Receipts (Kamai)
df_master['Deficit_Gap_Crores'] = df_master['Revenue_Expenditure'] - df_master['Total_Revenue_Receipts']

# 4. Save this perfect Master Data to your computer for Power BI / ML
df_master.to_csv('India_Budget_MasterData.csv', index=False)

print("\n--- Final Master Data Ready! ---")
display(df_master[['Year', 'Total_Revenue_Receipts', 'Revenue_Expenditure', 'Deficit_Gap_Crores']].head())
print("\nSuccess! Master data saved as 'India_Budget_MasterData.csv'.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load and clean the data
df_viz = pd.read_csv('India_Budget_MasterData.csv')
df_viz = df_viz.dropna(subset=['Year', 'Total_Revenue_Receipts', 'Revenue_Expenditure'])
df_viz['Year'] = df_viz['Year'].astype(str)
df_viz = df_viz.sort_values('Year')

# 2. Filter to keep ONLY the last 20 years
df_last_20 = df_viz.tail(20)

# 3. Set up the first chart (Line Chart)
sns.set_theme(style="whitegrid")
plt.figure(figsize=(12, 6))

# Plot Income and Expenditure
plt.plot(df_last_20['Year'], df_last_20['Total_Revenue_Receipts'], marker='o', label='Revenue (Income)', color='green', linewidth=2.5)
plt.plot(df_last_20['Year'], df_last_20['Revenue_Expenditure'], marker='s', label='Expenditure (Expense)', color='red', linewidth=2.5)

# Formatting
plt.title('Government Income vs Expenditure (Last 20 Years)', fontsize=16, fontweight='bold')
plt.xlabel('Financial Year', fontsize=12)
plt.ylabel('Amount (INR Crores)', fontsize=12)
plt.xticks(rotation=45) 
plt.legend()
plt.tight_layout()

# Show Chart 1
plt.show()
```

```python
# 1. Calculate YoY Growth for the Deficit
df_master['Deficit_YoY_Growth_%'] = df_master['Deficit_Gap_Crores'].pct_change(-1) * 100

# 2. Calculate what % of the Budget is the Gap (Deficit Ratio)
df_master['Deficit_to_Expense_Ratio_%'] = (df_master['Deficit_Gap_Crores'] / df_master['Revenue_Expenditure']) * 100

# 3. Display the most recent 5 years with these insights
print("--- Final Budgetary Insights (Recent Years) ---")
display(df_master[['Year', 'Deficit_Gap_Crores', 'Deficit_YoY_Growth_%', 'Deficit_to_Expense_Ratio_%']].head(5))

# 4. Quick Summary
latest_ratio = df_master['Deficit_to_Expense_Ratio_%'].iloc[0]
print(f"\nSummary Insight: In the latest budget year, the 'Gap' is {latest_ratio:.2f}% of the total revenue expenditure.")
```

```python
# 1. Calculate the absolute difference in Deficit from the previous year
# The .diff(-1) function calculates: (Current Year Deficit) - (Previous Year Deficit)
df_master['Deficit_Change_Value'] = df_master['Deficit_Gap_Crores'].diff(-1)

# 2. Calculate Year-on-Year (YoY) Growth Percentage
df_master['Deficit_YoY_Growth_%'] = df_master['Deficit_Gap_Crores'].pct_change(-1) * 100

# 3. Calculate the Deficit to Total Expense Ratio
# This shows what percentage of our total spending is actually "The Gap"
df_master['Deficit_Ratio_%'] = (df_master['Deficit_Gap_Crores'] / df_master['Revenue_Expenditure']) * 100

# 4. Filter and display the last 20 years of data
print("--- 20-Year Budgetary Trend Analysis ---")
# Selecting specific columns for a clear report
report_columns = [
    'Year', 
    'Deficit_Gap_Crores', 
    'Deficit_Change_Value', 
    'Deficit_YoY_Growth_%', 
    'Deficit_Ratio_%'
]

# Using .head(20) to see the most recent 20 years
analysis_report = df_master[report_columns].head(20)
display(analysis_report)

# 5. Brief Summary Insight
total_20yr_deficit = df_master['Deficit_Gap_Crores'].head(20).sum()
print(f"\nTotal Deficit over the last 20 years: ₹{total_20yr_deficit:,.2f} Crores")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression

print("Loading Data and Building Machine Learning Model...")

# --- THE FIX: Load the saved CSV file back into memory ---
df_master = pd.read_csv('India_Budget_MasterData.csv')
df_master = df_master.dropna(subset=['Year', 'Total_Revenue_Receipts', 'Revenue_Expenditure'])
df_master['Year'] = df_master['Year'].astype(str)
# --------------------------------------------------------

# --- 1. DATA PREPARATION FOR ML ---
df_ml = df_master.copy()

# Convert text years ('2025-26') into pure numbers (2025) so the ML model can understand them
df_ml['Numeric_Year'] = df_ml['Year'].str[:4].astype(int)

# Sort just to be safe and use the last 15 years for training
df_ml = df_ml.sort_values('Numeric_Year')
df_ml = df_ml.tail(15) 

# Define X (Input Feature = Year) and y (Target Output = Deficit)
X = df_ml[['Numeric_Year']] 
y = df_ml['Deficit_Gap_Crores']

# --- 2. TRAIN THE ML MODEL ---
model = LinearRegression()
model.fit(X, y) # The model is learning the mathematical relationship between Year and Deficit

# --- 3. MAKE FUTURE PREDICTIONS ---
# Let's predict for the next 3 financial years: 2026, 2027, and 2028
future_years = pd.DataFrame({'Numeric_Year': [2026, 2027, 2028]})
future_predictions = model.predict(future_years)

# Create a neat table to show the future results
df_future = future_years.copy()
df_future['Predicted_Deficit_Crores'] = future_predictions
df_future['Financial_Year'] = [f"{y}-{str(y+1)[-2:]}" for y in df_future['Numeric_Year']]

print("\n--- ML Predictions for Future Deficits ---")
display(df_future[['Financial_Year', 'Predicted_Deficit_Crores']])

# --- 4. VISUALIZE THE PREDICTION ---
plt.figure(figsize=(10, 6))
sns.set_theme(style="whitegrid")

# Plot the historical data as blue dots
plt.scatter(df_ml['Numeric_Year'], df_ml['Deficit_Gap_Crores'], color='blue', label='Actual Past Deficit', s=80)

# Plot the ML Regression Line (The mathematical trend)
all_years = pd.DataFrame({'Numeric_Year': np.arange(df_ml['Numeric_Year'].min(), 2029)})
plt.plot(all_years['Numeric_Year'], model.predict(all_years), color='red', linestyle='--', linewidth=2, label='ML Trend Line')

# Highlight the future predictions as golden stars
plt.scatter(future_years['Numeric_Year'], future_predictions, color='orange', label='Predicted Future Deficit', s=200, marker='*', zorder=5)

# Formatting the Chart
plt.title('Predicting India\'s Budget Deficit (Linear Regression)', fontsize=16, fontweight='bold')
plt.xlabel('Starting Year of Financial Budget', fontsize=12)
plt.ylabel('Deficit Amount (INR Crores)', fontsize=12)
plt.legend()
plt.tight_layout()
plt.show()

# --- 5. FINAL INSIGHT ---
pred_2027 = df_future[df_future['Numeric_Year'] == 2027]['Predicted_Deficit_Crores'].values[0]
print(f"\nML Insight: Based on the historical trend line, the Linear Regression model predicts that the budget deficit for the year 2027-28 will be approximately ₹{pred_2027:,.2f} Crores.")
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print("Upgrading to an Advanced Polynomial Regression Model...")

# --- 1. DATA PREPARATION ---
# Create a fresh copy of the master data
df_ml = df_master.copy()

# Extract the starting year as an integer (e.g., '2025-26' becomes 2025)
df_ml['Numeric_Year'] = df_ml['Year'].str[:4].astype(int)

# Use the last 15 years of data for training to capture the most relevant recent trends
df_ml = df_ml.sort_values('Numeric_Year').tail(15)

X = df_ml[['Numeric_Year']]
y = df_ml['Deficit_Gap_Crores']

# --- THE UPGRADE: Polynomial Features ---
# Using degree=3 allows the trend line to curve, capturing the COVID-19 spike and the recent recovery
poly = PolynomialFeatures(degree=3) 
X_poly = poly.fit_transform(X)

# --- 2. TRAIN THE ADVANCED MODEL ---
advanced_model = LinearRegression()
advanced_model.fit(X_poly, y)

# --- 3. PREDICT FUTURE TRENDS ---
# Predict the deficit for the upcoming financial years: 2026, 2027, and 2028
future_years = pd.DataFrame({'Numeric_Year': [2026, 2027, 2028]})
future_poly = poly.transform(future_years)
advanced_predictions = advanced_model.predict(future_poly)

# Store predictions in a clean dataframe for display
df_future_adv = future_years.copy()
df_future_adv['Predicted_Deficit_Crores'] = advanced_predictions
df_future_adv['Financial_Year'] = [f"{y}-{str(y+1)[-2:]}" for y in df_future_adv['Numeric_Year']]

print("\n--- Advanced ML Predictions (Polynomial) ---")
display(df_future_adv[['Financial_Year', 'Predicted_Deficit_Crores']])

# --- 4. VISUALIZE THE POLYNOMIAL TREND ---
plt.figure(figsize=(10, 6))
sns.set_theme(style="whitegrid")

# Plot the actual historical data points
plt.scatter(df_ml['Numeric_Year'], df_ml['Deficit_Gap_Crores'], color='blue', label='Actual Past Deficit', s=80)

# Generate a smooth curved line for the plot using the trained model
X_smooth = np.linspace(df_ml['Numeric_Year'].min(), 2028, 100).reshape(-1, 1)
X_smooth_poly = poly.transform(X_smooth)
y_smooth_pred = advanced_model.predict(X_smooth_poly)

# Plot the advanced Polynomial curve
plt.plot(X_smooth, y_smooth_pred, color='green', linestyle='-', linewidth=2.5, label='Advanced Trend (Polynomial Curve)')

# Highlight the future predictions
plt.scatter(future_years['Numeric_Year'], advanced_predictions, color='gold', label='Future Predictions', s=250, marker='*', zorder=5, edgecolor='black')

# Chart Formatting
plt.title('Predicting India\'s Budget Deficit (Advanced Polynomial Regression)', fontsize=16, fontweight='bold')
plt.xlabel('Starting Year of Financial Budget', fontsize=12)
plt.ylabel('Deficit Amount (INR Crores)', fontsize=12)
plt.legend()
plt.tight_layout()
plt.show()

# --- 5. FINAL INSIGHT ---
pred_2027_adv = df_future_adv[df_future_adv['Numeric_Year'] == 2027]['Predicted_Deficit_Crores'].values[0]
print(f"\n ML Insight: The Polynomial Regression model successfully captures the post-COVID recovery trend. The projected deficit for 2027-28 is now adjusted to approximately ₹{pred_2027_adv:,.2f} Crores.")
```
